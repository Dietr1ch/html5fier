```python
from IPython import display

import matplotlib
from matplotlib import pylab
import pandas as pd

from src.tags import TAGS, SEMANTIC_TAGS
TAGS = list(SEMANTIC_TAGS)
SEMANTIC_TAGS = list(SEMANTIC_TAGS)

matplotlib.style.use('ggplot')
matplotlib.rcParams['figure.figsize'] = (12, 10)

%matplotlib inline
```

```python
sites = pd.read_csv('dataset/sites_hist.csv')
```

```python
sites = sites[['site_url', 'total_tags', 'valid_tags', 'semantic_tags', 'semantic_tag_frac'] + list(SEMANTIC_TAGS)]
```

```python
sites[['semantic_tags'] + list(SEMANTIC_TAGS)].describe()
```

```python
sites.semantic_tag_frac.plot(kind='hist', figsize=(16, 12), logy=True, fontsize=12)
pylab.title(u'Histograma de Ocurrencia Relativa de Elementos Semánticos antes del Filtrado', fontsize=16)
pylab.xlabel(u'Proporción de elementos semanticos (%)', fontsize=12)
pylab.ylabel(u'Cantidad de sitios', fontsize=12)
```

```python
# Plot global element freq. v/s element freq. threshold.
thresholds = pylab.linspace(0., 1., 100)
global_freqs = []
for t in thresholds:
    freq = sites[sites.semantic_tag_frac > t].semantic_tags.sum()
    if freq == 0:
        break
    global_freqs.append(sites[sites.semantic_tag_frac > t].semantic_tags.sum())

element_freq = pd.Series(global_freqs, index=thresholds[:len(global_freqs)])
```

```python
element_freq.plot(figsize=(16, 12), fontsize=14, linewidth=2, yticks=pylab.arange(0, 250000, 15000))
pylab.title(u'Frecuencia de Elementos Semánticos a Través de Todos los Sitios', fontsize=18)
pylab.xlabel(u'Valor de corte en cada página (%)', fontsize=16)
pylab.ylabel(u'Cantidad total de elementos semánticos', fontsize=16)
pylab.grid(b=True, which='major', color=(.33, .33, .33), linestyle='--')
```

```python
# Filter at 0.05
sites = sites[sites.semantic_tag_frac > .05]
len(sites)
```

```python
sites[SEMANTIC_TAGS].sum().sort(inplace=False, ascending=False).plot(kind='bar', x=SEMANTIC_TAGS,
                                                                     figsize=(16, 12), fontsize=14,
                                                                     yticks=pylab.arange(0, 50000, 3000),
                                                                     rot=True)
pylab.title(u'Frecuencia de Elementos Semánticos a Través de Todos los Sitios por Tipo', fontsize=18)
pylab.xlabel(u'Elemento', fontsize=16)
pylab.ylabel(u'Frecuencia total', fontsize=16)
pylab.grid(b=True, which='major', color=(.33, .33, .33), linestyle='--')
pylab.grid(b=True, which='minor', color=(.33, .33, .33), linestyle='--')
```

```python
# Save filtered site url list.
with open('dataset/filtered_sites.txt', 'w') as url_file:
    for site_url in set(sites.site_url.values):
        url_file.write('{}\n'.format(site_url))
```
